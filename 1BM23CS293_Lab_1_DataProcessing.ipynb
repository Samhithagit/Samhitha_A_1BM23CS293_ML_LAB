```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# -----------------------------
# Load Dataset
# -----------------------------
adult = pd.read_csv("adult.csv")

# -----------------------------
# 1. Handling Missing Values
# -----------------------------
adult.replace("?", np.nan, inplace=True)

for col in adult.columns:
    if adult[col].dtype == "object":
        adult[col].fillna(adult[col].mode()[0], inplace=True)
    else:
        adult[col].fillna(adult[col].mean(), inplace=True)

print("Missing values handled")

# -----------------------------
# 2. Handling Categorical Data
# -----------------------------
adult = pd.get_dummies(adult, drop_first=True)

print("Categorical data converted to numeric")

# -----------------------------
# 3. Handling Outliers (IQR)
# -----------------------------
numeric_adult = adult.select_dtypes(include=['int64','float64'])

Q1 = numeric_adult.quantile(0.25)
Q3 = numeric_adult.quantile(0.75)

IQR = Q3 - Q1

adult = adult[~((numeric_adult < (Q1 - 1.5 * IQR)) |
                (numeric_adult > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Outliers removed")

# -----------------------------
# 4. Min-Max Normalization
# -----------------------------
minmax = MinMaxScaler()

adult_minmax = pd.DataFrame(
    minmax.fit_transform(adult),
    columns=adult.columns
)

print("Min-Max Scaling Output")
print(adult_minmax.head())

# -----------------------------
# 5. Standard Scaler
# -----------------------------
standard = StandardScaler()

adult_standard = pd.DataFrame(
    standard.fit_transform(adult),
    columns=adult.columns
)

print("Standard Scaling Output")
print(adult_standard.head())
```
