```python
# Linear Regression using Library

import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/canada_per_capita_income.csv")

# Display dataset
print("Dataset:")
print(df.head())

# Input and Output
X = df[['year']]
y = df['per capita income (US$)']

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict income for 2020
prediction = model.predict([[2020]])

print("\nUsing Library:")
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
print("Predicted Per Capita Income in 2020:", prediction[0])
```

```python
# Linear Regression without Library

import pandas as pd

# Load dataset
df = pd.read_csv("/content/canada_per_capita_income.csv")

# Convert to arrays
x = df['year'].values
y = df['per capita income (US$)'].values

n = len(x)

# Calculate required sums
sum_x = sum(x)
sum_y = sum(y)
sum_xy = sum(x*y)
sum_x2 = sum(x*x)

# Calculate slope (m)
m = (n*sum_xy - sum_x*sum_y) / (n*sum_x2 - sum_x**2)

# Calculate intercept (b)
b = (sum_y - m*sum_x) / n

# Predict income for 2020
x_test = 2020
prediction = m*x_test + b

print("\nWithout Library:")
print("Slope:", m)
print("Intercept:", b)
print("Predicted Per Capita Income in 2020:", prediction)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("hiring.csv")

print("Original Dataset:")
print(df)

# Step 1: Convert words to numbers
word_to_num = {
    'zero':0, 'one':1, 'two':2, 'three':3, 'four':4,
    'five':5, 'six':6, 'seven':7, 'eight':8,
    'nine':9, 'ten':10, 'eleven':11, 'twelve':12
}

df['experience'] = df['experience'].replace(word_to_num)

# Step 2: Fill missing values
df['experience'] = df['experience'].fillna(df['experience'].mean())
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean())

print("\nCleaned Dataset:")
print(df)

# Step 3: Input and Output
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

# Step 4: Train model
model = LinearRegression()
model.fit(X, y)

# Step 5: Predictions
salary1 = model.predict([[2, 9, 6]])
salary2 = model.predict([[12, 10, 10]])

print("\nUsing Library:")
print("Salary (2 yr, 9 test, 6 interview):", salary1[0])
print("Salary (12 yr, 10 test, 10 interview):", salary2[0])

```

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("hiring.csv")

# Convert words to numbers
word_to_num = {
    'zero':0, 'one':1, 'two':2, 'three':3, 'four':4,
    'five':5, 'six':6, 'seven':7, 'eight':8,
    'nine':9, 'ten':10, 'eleven':11, 'twelve':12
}

df['experience'] = df['experience'].replace(word_to_num)

# Fill missing values
df['experience'] = df['experience'].fillna(df['experience'].mean())
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean())

# Input and output
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']].values
Y = df['salary($)'].values

# Add column of 1s
ones = np.ones((X.shape[0], 1))
X = np.concatenate((ones, X), axis=1)

# Normal equation
B = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)

# Predictions
x1 = np.array([1, 2, 9, 6])
salary1 = x1.dot(B)

x2 = np.array([1, 12, 10, 10])
salary2 = x2.dot(B)

print("\nWithout Library:")
print("Salary (2 yr, 9 test, 6 interview):", salary1)
print("Salary (12 yr, 10 test, 10 interview):", salary2)
```

```python
# Multiple Linear Regression using Library

import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("1000_Companies.csv")

print("Dataset:")
print(df.head())

# Convert categorical variable (State) into dummy variables
df = pd.get_dummies(df, columns=['State'], drop_first=True)

print("\nDataset after encoding:")
print(df.head())

# Input variables
X = df.drop('Profit', axis=1)

# Output variable
y = df['Profit']

# Train model
model = LinearRegression()
model.fit(X, y)

# Prediction for given values
# Florida → State_Florida = 1, State_New York = 0
prediction = model.predict([[91694.48, 515841.3, 11931.24, 1, 0]])

print("\nUsing Library:")
print("Predicted Profit:", prediction[0])
```

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("1000_Companies.csv")

# Convert State column into dummy variables
df = pd.get_dummies(df, columns=['State'], drop_first=True)

# Convert all columns to float (VERY IMPORTANT FIX)
df = df.astype(float)

# Input variables
X = df.drop('Profit', axis=1).values

# Output variable
Y = df['Profit'].values

# Add intercept column (1s)
ones = np.ones((X.shape[0], 1))
X = np.concatenate((ones, X), axis=1)

# Apply Normal Equation
B = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)

print("Coefficients:", B)

# Prediction for given input
# Florida = State_Florida=1, State_New York=0
x_test = np.array([1, 91694.48, 515841.3, 11931.24, 1, 0], dtype=float)

prediction = x_test.dot(B)

print("\nWithout Library:")
print("Predicted Profit:", prediction)

```
